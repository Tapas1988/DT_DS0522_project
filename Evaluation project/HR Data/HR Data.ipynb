```python
# for getting iqr

q3=dt.quantile(0.75)
q1=dt.quantile(0.25)
iqr=q3-q1
```

```python
#  handle Outlir MonthlyIncome column
higher_MonthlyIncome=q3.MonthlyIncome+(1.5*iqr.MonthlyIncome)  # find hiher side outlier
print('higher side   >',higher_MonthlyIncome)

index=np.where(dt['MonthlyIncome']>higher_MonthlyIncome)     # find index number of higher outlier
dt=dt.drop(dt.index[index])          # drop the index
dt.reset_index()                     # reset index of data
dt.shape                             # shape the data
```
